```python
from crewai import Agent, Crew, Process, Task
from crewai_tools import SerperDevTool

# ğŸ” Web search tool for real-time finance research
search_tool = SerperDevTool()

# ğŸ§  Define the Research agent
researcher = Agent(
    role="Financial Research Analyst",
    goal="Gather up-to-date financial data, trends, and news for the target companies or markets",
    backstory="""
        You are a Senior Financial Research Analyst.
        Use the web search tool to find the most relevant financial updates,
        stock performance summaries, P/E ratios, major news, market outlooks, etc.
    """,
    verbose=True,
    allow_delegation=False,
    tools=[search_tool],  # enables web search
)

# âœï¸ Define the Writer agent
writer = Agent(
    role="Financial Report Writer",
    goal="Compile and summarize financial research into clear, actionable insights",
    backstory="""
        You are an experienced financial content writer.
        Turn raw research into clear analysis with bullet points and summaries.
    """,
    verbose=True,
    allow_delegation=True,  
)

# ğŸ“‹ Task: Financial research
task1 = Task(
    description="""
        Perform financial research on the specified topic or ticker symbols.
        Find the latest performance, key ratios, major news, and analyst sentiment.
    """,
    expected_output="Detailed financial research summary with web search findings",
    agent=researcher,
)

# ğŸ“„ Task: Generate summary report
task2 = Task(
    description="""
        Based on the research above, write a cohesive financial analysis report.
        Present key takeaways, market implications, and simplified insights for the user.
    """,
    expected_output="A polished financial analysis report",
    agent=writer,
)

# ğŸ§© Create and run the crew sequentially
crew = Crew(
    agents=[researcher, writer],
    tasks=[task1, task2],
    verbose=1,
    process=Process.sequential,
)
```

```python
result = crew.kickoff()

print("\n=== Financial Analysis Output ===")
print(result)
```
